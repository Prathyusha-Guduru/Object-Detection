```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('grid.png')
plt.imshow(img)
```

```python
found,corners = cv2.findChessboardCorners(img,(7,7))
if found == True:
    print("Corners found")
else:
    print("Not found")
```

```python
new_img = cv2.drawChessboardCorners(img,(7,7),corners,found)
```

```python
plt.imshow(new_img)
```
