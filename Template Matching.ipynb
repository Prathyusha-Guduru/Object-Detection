# Template Matching

It is just finding the location of a template image(sub-set) in a larger image(super-set)
One has to note that the the exact template should be present in the larger image without any subtle differences for the detection to be succesfull.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
large_img = cv2.imread('Large_image.jpg')
large_img = cv2.cvtColor(large_img,cv2.COLOR_BGR2RGB)
plt.imshow(large_img)
```

```python
template = cv2.imread('Template.jpg')
template = cv2.cvtColor(template,cv2.COLOR_BGR2RGB)
plt.imshow(template)
```

```python
c,width,height = template.shape[::-1]
```

```python
temp_match_methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for x in temp_match_methods:
    
    img = large_img.copy()
    method = eval(x)
    result = cv2.matchTemplate(img,template,method)
    _,__,min_loc,max_loc = cv2.minMaxLoc(result)
    if method == cv2.TM_SQDIFF or method == cv2.TM_SQDIFF_NORMED:
        top_right = min_loc
    else:
        top_right = max_loc
    bottom_left = (top_right[0]+width,top_right[1]+height)
    
    cv2.rectangle(img,top_right,bottom_left,(255,0,0),10)
     # Plot the Images
    plt.subplot(121)
    plt.imshow(result)
    plt.title('Heatmap')
    
    plt.subplot(122)
    plt.imshow(img)
    plt.title('Detected Point')
    
    plt.suptitle(x)
    plt.show()
    
    print('\n')
    print('\n')
```
