# Canny Edge Detector

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('sammy_face.jpg')
plt.imshow(img)
```

```python
edges = cv2.Canny(img,127,127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(img,0,127)
plt.imshow(edges)
```

lower = int(max(0,0.7*np.median(img)))

**Lower threshold is either 0 or 70% of the median value, whicever is higher**

**Upper threshold is either 255 or 130% of the median value (or 30% above the median value),whichever is lower**

```python
upper = int(min(255,1.3*np.median(img)))
upper
```
